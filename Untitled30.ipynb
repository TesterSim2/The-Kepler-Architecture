```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import triton
import triton.language as tl
import math

# ==========================================
# 1. TRITON KERNEL: SINKHORN-KNOPP
# ==========================================
# This kernel performs the iterative manifold projection.
# It forces the matrix to be Doubly Stochastic (Row Sum = 1, Col Sum = 1).
# We hardcode max_iter=20 as per DeepSeek paper.

@triton.jit
def sinkhorn_fwd_kernel(
    logits_ptr,      # [Total_Tokens, N, N]
    output_ptr,      # [Total_Tokens, N, N]
    stride_b,        # Stride for batch dimension
    stride_n1,       # Stride for first N
    stride_n2,       # Stride for second N
    N: tl.constexpr, # Expansion rate (e.g., 4)
    ITER: tl.constexpr,
    BLOCK_SIZE: tl.constexpr
):
    # Map program ID to the specific matrix in the batch
    pid = tl.program_id(0)

    # Offsets for the current matrix
    # The matrix is size NxN.
    # We load the entire NxN matrix into registers for a single thread/program if N is small.
    # Since N=4 is tiny, each program instance handles ONE matrix (one token's stream mixer).

    offs_n = tl.arange(0, N)

    # Pointers to the matrix start
    # We need to load N*N elements.
    # We use a grid logic: row indices [0..N], col indices [0..N]

    # Triton logic for small matrix operations:
    # We can treat the pointer arithmetic manually.
    base_ptr = logits_ptr + pid * stride_b

    # Load the matrix
    # Shape [N, N]
    # We use broadcasting to create the grid of pointers
    ptrs = base_ptr + (offs_n[:, None] * stride_n1) + (offs_n[None, :] * stride_n2)
    vals = tl.load(ptrs)

    # 1. Exponentiate (Ensure positivity)
    # Numerical stability hack: subtract max to prevent inf
    max_val = tl.max(vals)
    vals = tl.exp(vals - max_val)

    # 2. Sinkhorn Iterations
    # Loop unrolling for T_max = 20
    for i in range(ITER):
        # Row Normalization
        # Sum across columns (dim=1)
        row_sum = tl.sum(vals, axis=1)
        # Avoid division by zero
        vals = vals / (row_sum[:, None] + 1e-6)

        # Column Normalization
        # Sum across rows (dim=0)
        col_sum = tl.sum(vals, axis=0)
        vals = vals / (col_sum[None, :] + 1e-6)

    # 3. Store result
    out_base_ptr = output_ptr + pid * stride_b
    out_ptrs = out_base_ptr + (offs_n[:, None] * stride_n1) + (offs_n[None, :] * stride_n2)
    tl.store(out_ptrs, vals)

# ==========================================
# 2. AUTOGRAD WRAPPER (Forward-Triton, Backward-Recompute)
# ==========================================

class SinkhornFunction(torch.autograd.Function):
    @staticmethod
    def forward(ctx, logits, n_iters=20):
        # logits shape: [Batch*Seq, N, N]
        # We flatten batch and seq for the kernel
        original_shape = logits.shape
        logits_flat = logits.contiguous().view(-1, original_shape[-2], original_shape[-1])

        B_total, N, _ = logits_flat.shape

        # FIX: Use empty_like, not empty_as
        output = torch.empty_like(logits_flat)

        # Launch Triton Kernel
        # We ensure strides are from the contiguous flattened tensor
        grid = (B_total, )
        sinkhorn_fwd_kernel[grid](
            logits_flat, output,
            logits_flat.stride(0), logits_flat.stride(1), logits_flat.stride(2),
            N=N, ITER=n_iters, BLOCK_SIZE=32
        )

        # Save logits for recomputing backward pass
        ctx.save_for_backward(logits)
        ctx.n_iters = n_iters

        return output.view(original_shape)

    @staticmethod
    def backward(ctx, grad_output):
        """
        Paper Section 4.3.2: Recomputing.
        """
        logits, = ctx.saved_tensors
        n_iters = ctx.n_iters

        # Enable grad for the recomputation
        with torch.enable_grad():
            # Detach and require grad to track the recomputation graph
            x = logits.detach().requires_grad_(True)

            # --- PyTorch Re-implementation of Sinkhorn (Matches Triton Logic) ---
            # 1. Exp
            # Subtract max for stability (detached)
            x_stable = x - x.max(dim=-1, keepdim=True)[0].max(dim=-2, keepdim=True)[0].detach()
            M = torch.exp(x_stable)

            # 2. Iterations
            for _ in range(n_iters):
                # Row Norm
                M = M / (M.sum(dim=-1, keepdim=True) + 1e-6)
                # Col Norm
                M = M / (M.sum(dim=-2, keepdim=True) + 1e-6)

            # 3. Backward from the output gradients
            # We explicitly compute gradients of M w.r.t x
            # grad_output is dL/dM. We want dL/dx.
            # torch.autograd.backward allows us to feed dL/dM into the graph
            torch.autograd.backward(M, grad_output)

        return x.grad, None

def sinkhorn_knopp(logits, n_iters=20):
    return SinkhornFunction.apply(logits, n_iters)

# ==========================================
# 3. mHC LAYER IMPLEMENTATION
# ==========================================

class ManifoldConstrainedHyperConnection(nn.Module):
    def __init__(self, d_model, expansion_rate=4, init_std=0.02):
        super().__init__()
        self.d = d_model
        self.n = expansion_rate

        # Input dimension is n * d (flattened stream)
        self.input_dim = self.n * self.d

        # Dynamic Predictor (Eq 7)
        # We need to predict:
        # 1. H_pre  (1 x n) -> n parameters
        # 2. H_post (1 x n) -> n parameters
        # 3. H_res  (n x n) -> n*n parameters
        total_pred_params = self.n + self.n + (self.n * self.n)

        # Note: Paper applies RMSNorm to the input stream before prediction
        self.stream_norm = nn.RMSNorm(self.input_dim)

        # The Linear Predictor
        self.predictor = nn.Linear(self.input_dim, total_pred_params)

        # Learnable Gating Factors (alpha) and Biases (b)
        # We merge these into the linear layer logic or keep separate as per Eq 5/7.
        # Paper says: H = alpha * activation(Linear(x)) + b
        # We will use the Linear layer for the projection and apply alpha/bias explicitly.

        self.alpha_pre = nn.Parameter(torch.tensor(0.01))
        self.alpha_post = nn.Parameter(torch.tensor(0.01))
        self.alpha_res = nn.Parameter(torch.tensor(0.01))

        # Biases are implicitly handled if we enable bias in Linear,
        # but the paper has explicit b_pre, b_post, b_res.
        # We will let the Linear layer handle the 'projection' and add explicit bias if needed,
        # but standard Linear(bias=True) covers this mathematically.

        # --- INITIALIZATION (Crucial for Stability) ---
        # 1. Initialize predictor weights to 0 so we rely on biases initially
        nn.init.zeros_(self.predictor.weight)
        nn.init.zeros_(self.predictor.bias)

        # 2. Initialize H_res bias to approximate Identity after Sinkhorn
        # Sinkhorn(exp(Identity * Large)) -> Identity
        # We identify the slice of the bias corresponding to H_res
        res_start = 2 * self.n
        res_end = res_start + (self.n * self.n)

        with torch.no_grad():
            # Create an identity-like pattern for the bias
            identity_bias = torch.eye(self.n).flatten() * 10.0 # Large value favors diagonal
            self.predictor.bias[res_start:res_end] = identity_bias

    def forward(self, X_stream, layer_fn):
        """
        X_stream: [Batch, Seq, n, d]
        layer_fn: function (e.g., Attention + FFN) taking [B, S, d] -> [B, S, d]
        """
        B, S, n, d = X_stream.shape

        # 1. Flatten Stream for Prediction
        X_flat = X_stream.view(B, S, -1) # [B, S, n*d]

        # 2. Norm & Predict
        # Eq 7: x_tilde = RMSNorm(x)
        # Eq 7: H_raw = Linear(x_tilde)
        X_norm = self.stream_norm(X_flat)
        logits = self.predictor(X_norm) # [B, S, n + n + n*n]

        # 3. Split Logits
        # H_pre:  [0 : n]
        # H_post: [n : 2n]
        # H_res:  [2n : end]
        h_pre_raw = logits[..., :n]
        h_post_raw = logits[..., n:2*n]
        h_res_raw = logits[..., 2*n:].view(B, S, n, n)

        # 4. Apply Manifold Constraints (Eq 8)

        # H_pre = Sigmoid (Input Gating)
        # Note: Applying alpha scaling
        H_pre = torch.sigmoid(h_pre_raw) * self.alpha_pre

        # H_post = 2 * Sigmoid (Output Gating)
        H_post = 2.0 * torch.sigmoid(h_post_raw) * self.alpha_post

        # H_res = Sinkhorn (Residual Mixing)
        # Note: We apply alpha scaling to the logits *before* or the result *after*?
        # Eq 5 says: H_res = alpha * tanh(...) + b.
        # Eq 8 says: H_res = Sinkhorn(H_tilde_res).
        # We follow Eq 8 which constrains the FINAL matrix.
        # The logits entering Sinkhorn should be the raw predictions.
        H_res = sinkhorn_knopp(h_res_raw * self.alpha_res)

        # 5. Compute Layer Input (Collapse Stream)
        # X_layer_in = Sum_over_n( H_pre[i] * X[i] )
        # Shapes: H_pre [B, S, n], X_stream [B, S, n, d]
        X_layer_in = torch.einsum('bsn, bsnd -> bsd', H_pre, X_stream)

        # 6. Execute Layer (The "Black Box" Transformer Block)
        X_layer_out = layer_fn(X_layer_in) # [B, S, d]

        # 7. Update Stream (Eq 3 / 4)
        # X_{l+1} = (H_res @ X_l) + (H_post * F(X))

        # Term A: Mix existing streams
        # H_res [B, S, n, n] @ X_stream [B, S, n, d] -> [B, S, n, d]
        X_mixed = torch.einsum('bsnm, bsmd -> bsnd', H_res, X_stream)

        # Term B: Add new information
        # H_post [B, S, n] * X_layer_out [B, S, d] -> [B, S, n, d] (Broadcasting)
        X_update = torch.einsum('bsn, bsd -> bsnd', H_post, X_layer_out)

        X_next = X_mixed + X_update

        return X_next

# ==========================================
# 4. VERIFICATION SCRIPT
# ==========================================

def test_mhc_stability():
    print("Initializing mHC Test on GPU...")
    device = torch.device("cuda:0")
    torch.manual_seed(42)

    # Configuration
    B, S, N, D = 2, 128, 4, 64
    expansion_rate = N

    model = ManifoldConstrainedHyperConnection(D, expansion_rate).to(device)

    # Create Dummy Stream
    X_stream = torch.randn(B, S, N, D, device=device, requires_grad=True)

    # Dummy Layer Function (e.g., Identity for pure stability check)
    layer_fn = lambda x: x

    print("\n--- Forward Pass ---")
    X_out = model(X_stream, layer_fn)
    print(f"Input Shape: {X_stream.shape}")
    print(f"Output Shape: {X_out.shape}")

    # Verify Sinkhorn Property
    # We need to hook into the forward to grab H_res or infer it
    # Let's inspect the gradients to ensure flow
    loss = X_out.mean()
    print(f"Loss: {loss.item():.4f}")

    print("\n--- Backward Pass ---")
    loss.backward()
    print(f"Input Gradients Norm: {X_stream.grad.norm().item():.4f}")
    print("Backward pass successful (Recomputing verified).")

    # Verify Manifold Constraint (Doubly Stochastic)
    print("\n--- Manifold Constraint Verification ---")
    # We manually run the predictor to inspect H_res
    with torch.no_grad():
        X_flat = X_stream.view(B, S, -1)
        logits = model.predictor(model.stream_norm(X_flat))
        h_res_raw = logits[..., 2*N:].view(B, S, N, N)

        # Run Sinkhorn
        H_res = sinkhorn_knopp(h_res_raw * model.alpha_res)

        # Check Row Sums
        row_sums = H_res.sum(dim=-1)
        col_sums = H_res.sum(dim=-2)

        print(f"Target Sum: 1.0")
        print(f"Max Row Sum: {row_sums.max().item():.6f}")
        print(f"Min Row Sum: {row_sums.min().item():.6f}")
        print(f"Max Col Sum: {col_sums.max().item():.6f}")
        print(f"Min Col Sum: {col_sums.min().item():.6f}")

        if torch.allclose(row_sums, torch.ones_like(row_sums), atol=1e-5):
            print("SUCCESS: Matrix is Doubly Stochastic.")
        else:
            print("FAILURE: Matrix is not Doubly Stochastic.")

if __name__ == "__main__":
    test_mhc_stability()
```

```python
class MHCTransformerBlock(nn.Module):
    def __init__(self, d_model, n_heads, expansion_rate=4):
        super().__init__()
        self.d_model = d_model
        self.n = expansion_rate

        # 1. The Compute Engines (Standard)
        # We perform Attention and MLP on the collapsed (projected) input
        self.attn = nn.MultiheadAttention(d_model, n_heads, batch_first=True)
        self.mlp = nn.Sequential(
            nn.Linear(d_model, 4 * d_model),
            nn.GELU(),
            nn.Linear(4 * d_model, d_model)
        )

        # 2. The Norms (Applied inside the layer function wrapper usually,
        # but mHC does its own norm on the stream.
        # We usually norm the *input* to Attn/MLP for Pre-Norm stability.)
        self.norm1 = nn.RMSNorm(d_model)
        self.norm2 = nn.RMSNorm(d_model)

        # 3. The Hyper-Connections (Replacing the Residuals)
        # One mHC to manage the Attention residual
        self.mhc_attn = ManifoldConstrainedHyperConnection(d_model, expansion_rate)

        # One mHC to manage the MLP residual
        self.mhc_mlp = ManifoldConstrainedHyperConnection(d_model, expansion_rate)

    def forward(self, X_stream, mask=None):
        """
        Input: X_stream [B, S, n, d]  <-- The 'Hyper' Residual Stream
        Output: X_stream [B, S, n, d] <-- The updated stream
        """

        # --- Sub-Layer 1: Attention ---
        # Define the function to inject into mHC
        def attn_wrapper(x):
            # x is [B, S, d] (collapsed from stream)
            normed = self.norm1(x)
            # Standard Attention
            attn_out, _ = self.attn(normed, normed, normed, attn_mask=mask, is_causal=True)
            return attn_out

        # Pass stream through mHC.
        # mHC handles: Collapse -> Attn_Wrapper -> Expand -> Mix -> Add
        X_stream = self.mhc_attn(X_stream, attn_wrapper)

        # --- Sub-Layer 2: MLP ---
        def mlp_wrapper(x):
            normed = self.norm2(x)
            return self.mlp(normed)

        X_stream = self.mhc_mlp(X_stream, mlp_wrapper)

        return X_stream
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from transformers import GPT2Tokenizer
from datasets import load_dataset
import math
import time
import matplotlib.pyplot as plt

# ==========================================
# 1. LLAMA 3 COMPONENTS (RoPE, SwiGLU)
# ==========================================

class LlamaRMSNorm(nn.Module):
    def __init__(self, hidden_size, eps=1e-6):
        super().__init__()
        self.weight = nn.Parameter(torch.ones(hidden_size))
        self.eps = eps

    def forward(self, hidden_states):
        input_dtype = hidden_states.dtype
        hidden_states = hidden_states.to(torch.float32)
        variance = hidden_states.pow(2).mean(-1, keepdim=True)
        hidden_states = hidden_states * torch.rsqrt(variance + self.eps)
        return self.weight * hidden_states.to(input_dtype)

class RotaryEmbedding(nn.Module):
    def __init__(self, dim, max_seq_len=2048):
        super().__init__()
        inv_freq = 1.0 / (10000 ** (torch.arange(0, dim, 2).float() / dim))
        t = torch.arange(max_seq_len).float()
        freqs = torch.outer(t, inv_freq)
        emb = torch.cat((freqs, freqs), dim=-1)
        self.register_buffer("cos_cached", emb.cos()[None, None, :, :], persistent=False)
        self.register_buffer("sin_cached", emb.sin()[None, None, :, :], persistent=False)

    def forward(self, x, seq_len=None):
        # x: [batch, heads, seq_len, head_dim]
        return self.cos_cached[:, :, :seq_len, ...], self.sin_cached[:, :, :seq_len, ...]

def rotate_half(x):
    x1, x2 = x.chunk(2, dim=-1)
    return torch.cat((-x2, x1), dim=-1)

def apply_rotary_pos_emb(x, cos, sin):
    return (x * cos) + (rotate_half(x) * sin)

class LlamaAttention(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.n_heads = config['n_heads']
        self.head_dim = config['d_model'] // self.n_heads
        self.wq = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.wk = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.wv = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.wo = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.rope = RotaryEmbedding(self.head_dim, config['context_len'])

    def forward(self, x):
        B, S, D = x.shape
        # Projections
        xq = self.wq(x).view(B, S, self.n_heads, self.head_dim).transpose(1, 2)
        xk = self.wk(x).view(B, S, self.n_heads, self.head_dim).transpose(1, 2)
        xv = self.wv(x).view(B, S, self.n_heads, self.head_dim).transpose(1, 2)

        # RoPE
        cos, sin = self.rope(xv, seq_len=S)
        xq = apply_rotary_pos_emb(xq, cos, sin)
        xk = apply_rotary_pos_emb(xk, cos, sin)

        # Attention
        out = F.scaled_dot_product_attention(xq, xk, xv, is_causal=True)

        # Output Projection
        out = out.transpose(1, 2).contiguous().view(B, S, D)
        return self.wo(out)

class LlamaSwiGLUMLP(nn.Module):
    def __init__(self, config):
        super().__init__()
        hidden_dim = 4 * config['d_model']
        hidden_dim = int(2 * hidden_dim / 3) # Llama style reduction
        self.w1 = nn.Linear(config['d_model'], hidden_dim, bias=False) # Gate
        self.w2 = nn.Linear(config['d_model'], hidden_dim, bias=False) # Value
        self.w3 = nn.Linear(hidden_dim, config['d_model'], bias=False) # Down

    def forward(self, x):
        return self.w3(F.silu(self.w1(x)) * self.w2(x))

# ==========================================
# 2. THE NANO-LLAMA mHC MODEL
# ==========================================

class MHCLlamaBlock(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.norm1 = LlamaRMSNorm(config['d_model'])
        self.attn = LlamaAttention(config)

        self.norm2 = LlamaRMSNorm(config['d_model'])
        self.mlp = LlamaSwiGLUMLP(config)

        # The Rigorous mHC Modules
        self.mhc_attn = ManifoldConstrainedHyperConnection(config['d_model'], config['expansion_rate'])
        self.mhc_mlp = ManifoldConstrainedHyperConnection(config['d_model'], config['expansion_rate'])

    def forward(self, X_stream):
        # 1. Attention Sub-layer
        def attn_wrapper(x):
            return self.attn(self.norm1(x))
        X_stream = self.mhc_attn(X_stream, attn_wrapper)

        # 2. MLP Sub-layer
        def mlp_wrapper(x):
            return self.mlp(self.norm2(x))
        X_stream = self.mhc_mlp(X_stream, mlp_wrapper)

        return X_stream

class NanoLlamaMHC(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.config = config
        self.n = config['expansion_rate']

        self.token_emb = nn.Embedding(config['vocab_size'], config['d_model'])

        # We replicate input embeddings n times to fill the stream.
        self.layers = nn.ModuleList([MHCLlamaBlock(config) for _ in range(config['n_layers'])])

        self.norm_final = LlamaRMSNorm(config['d_model'])
        self.lm_head = nn.Linear(config['d_model'], config['vocab_size'], bias=False)

        # Weight tying
        self.lm_head.weight = self.token_emb.weight

    def forward(self, idx, targets=None):
        B, S = idx.shape

        # 1. Embedding
        x = self.token_emb(idx) # [B, S, D]

        # 2. Expand to Stream [B, S, n, D]
        X_stream = x.unsqueeze(2).expand(-1, -1, self.n, -1).clone()

        # 3. Layers
        for layer in self.layers:
            X_stream = layer(X_stream)

        # 4. Collapse Stream
        # Average pool the streams to restore identity
        x_final = X_stream.mean(dim=2)

        # 5. Output
        x_final = self.norm_final(x_final)
        logits = self.lm_head(x_final)

        loss = None
        if targets is not None:
            # FIX: Use .reshape() instead of .view() to handle non-contiguous slices
            loss = F.cross_entropy(logits.reshape(-1, logits.size(-1)), targets.reshape(-1))

        return logits, loss
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import triton
import triton.language as tl
import math

# ==========================================
# 1. TRITON KERNEL: SINKHORN-KNOPP
# ==========================================
@triton.jit
def sinkhorn_fwd_kernel(
    logits_ptr,      # [Total_Tokens, N, N]
    output_ptr,      # [Total_Tokens, N, N]
    stride_b,        # Stride for batch dimension
    stride_n1,       # Stride for first N
    stride_n2,       # Stride for second N
    N: tl.constexpr, # Expansion rate (e.g., 4)
    ITER: tl.constexpr,
    BLOCK_SIZE: tl.constexpr
):
    pid = tl.program_id(0)
    offs_n = tl.arange(0, N)

    base_ptr = logits_ptr + pid * stride_b
    ptrs = base_ptr + (offs_n[:, None] * stride_n1) + (offs_n[None, :] * stride_n2)
    vals = tl.load(ptrs)

    # Stability: subtract max
    max_val = tl.max(vals)
    vals = tl.exp(vals - max_val)

    # Sinkhorn Iterations
    for i in range(ITER):
        # Row Norm
        row_sum = tl.sum(vals, axis=1)
        vals = vals / (row_sum[:, None] + 1e-6)
        # Column Norm
        col_sum = tl.sum(vals, axis=0)
        vals = vals / (col_sum[None, :] + 1e-6)

    out_base_ptr = output_ptr + pid * stride_b
    out_ptrs = out_base_ptr + (offs_n[:, None] * stride_n1) + (offs_n[None, :] * stride_n2)
    tl.store(out_ptrs, vals)

# ==========================================
# 2. AUTOGRAD WRAPPER
# ==========================================
class SinkhornFunction(torch.autograd.Function):
    @staticmethod
    def forward(ctx, logits, n_iters=20):
        original_shape = logits.shape
        # Contiguous view is required for Triton strides
        logits_flat = logits.contiguous().view(-1, original_shape[-2], original_shape[-1])
        B_total, N, _ = logits_flat.shape

        # Allocate output with same attributes as input
        output = torch.empty_like(logits_flat)

        grid = (B_total, )
        sinkhorn_fwd_kernel[grid](
            logits_flat, output,
            logits_flat.stride(0), logits_flat.stride(1), logits_flat.stride(2),
            N=N, ITER=n_iters, BLOCK_SIZE=32
        )

        ctx.save_for_backward(logits)
        ctx.n_iters = n_iters
        return output.view(original_shape)

    @staticmethod
    def backward(ctx, grad_output):
        logits, = ctx.saved_tensors
        n_iters = ctx.n_iters

        with torch.enable_grad():
            x = logits.detach().requires_grad_(True)
            # Recompute logic
            x_stable = x - x.max(dim=-1, keepdim=True)[0].max(dim=-2, keepdim=True)[0].detach()
            M = torch.exp(x_stable)
            for _ in range(n_iters):
                M = M / (M.sum(dim=-1, keepdim=True) + 1e-6)
                M = M / (M.sum(dim=-2, keepdim=True) + 1e-6)

            torch.autograd.backward(M, grad_output)

        return x.grad, None

def sinkhorn_knopp(logits, n_iters=20):
    return SinkhornFunction.apply(logits, n_iters)

# ==========================================
# 3. mHC LAYER DEFINITION
# ==========================================
class ManifoldConstrainedHyperConnection(nn.Module):
    def __init__(self, d_model, expansion_rate=4, init_std=0.02):
        super().__init__()
        self.d = d_model
        self.n = expansion_rate
        self.input_dim = self.n * self.d

        # Predictor parameters: H_pre (n) + H_post (n) + H_res (n*n)
        total_pred_params = self.n + self.n + (self.n * self.n)

        self.stream_norm = nn.RMSNorm(self.input_dim)
        self.predictor = nn.Linear(self.input_dim, total_pred_params)

        # Gating factors
        self.alpha_pre = nn.Parameter(torch.tensor(0.01))
        self.alpha_post = nn.Parameter(torch.tensor(0.01))
        self.alpha_res = nn.Parameter(torch.tensor(0.01))

        # Initialization
        nn.init.zeros_(self.predictor.weight)
        nn.init.zeros_(self.predictor.bias)

        # Identity Initialization for H_res
        res_start = 2 * self.n
        res_end = res_start + (self.n * self.n)
        with torch.no_grad():
            # Large diagonal bias ensures Softmax/Sinkhorn approximates Identity
            identity_bias = torch.eye(self.n).flatten() * 10.0
            self.predictor.bias[res_start:res_end] = identity_bias

    def forward(self, X_stream, layer_fn):
        B, S, n, d = X_stream.shape
        X_flat = X_stream.view(B, S, -1)

        X_norm = self.stream_norm(X_flat)
        logits = self.predictor(X_norm)

        h_pre_raw = logits[..., :n]
        h_post_raw = logits[..., n:2*n]
        h_res_raw = logits[..., 2*n:].view(B, S, n, n)

        H_pre = torch.sigmoid(h_pre_raw) * self.alpha_pre
        H_post = 2.0 * torch.sigmoid(h_post_raw) * self.alpha_post
        H_res = sinkhorn_knopp(h_res_raw * self.alpha_res)

        # Collapse
        X_layer_in = torch.einsum('bsn, bsnd -> bsd', H_pre, X_stream)

        # Layer
        X_layer_out = layer_fn(X_layer_in)

        # Expand/Update
        X_mixed = torch.einsum('bsnm, bsmd -> bsnd', H_res, X_stream)
        X_update = torch.einsum('bsn, bsd -> bsnd', H_post, X_layer_out)

        return X_mixed + X_update
```

```python
# ==============================================================================
# RIGOROUS IMPLEMENTATION: MANIFOLD-CONSTRAINED HYPER-CONNECTIONS (mHC)
#
# Paper Reference: "mHC: Manifold-Constrained Hyper-Connections" (DeepSeek-AI)
# Implemented by: Norm (Senior Researcher) & User
# ==============================================================================

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torch.optim.lr_scheduler import LambdaLR
import triton
import triton.language as tl
import math
import time
import matplotlib.pyplot as plt
from transformers import GPT2Tokenizer
from datasets import load_dataset

# ==========================================
# 1. TRITON KERNEL: SINKHORN-KNOPP
# ==========================================
# This kernel projects the residual transition matrix onto the Birkhoff Polytope
# (Doubly Stochastic Matrix) using iterative row/column normalization.

@triton.jit
def sinkhorn_fwd_kernel(
    logits_ptr,      # [Total_Tokens, N, N]
    output_ptr,      # [Total_Tokens, N, N]
    stride_b,        # Stride for batch dimension
    stride_n1,       # Stride for first N
    stride_n2,       # Stride for second N
    N: tl.constexpr, # Expansion rate (e.g., 4)
    ITER: tl.constexpr,
    BLOCK_SIZE: tl.constexpr
):
    pid = tl.program_id(0)
    offs_n = tl.arange(0, N)

    # Calculate pointer offsets
    base_ptr = logits_ptr + pid * stride_b
    ptrs = base_ptr + (offs_n[:, None] * stride_n1) + (offs_n[None, :] * stride_n2)

    # Load Logits
    vals = tl.load(ptrs)

    # 1. Exponentiate for positivity (Numerical Stability: subtract max)
    max_val = tl.max(vals)
    vals = tl.exp(vals - max_val)

    # 2. Sinkhorn Iterations (In-Register)
    for i in range(ITER):
        # Row Normalization
        row_sum = tl.sum(vals, axis=1)
        vals = vals / (row_sum[:, None] + 1e-6)

        # Column Normalization
        col_sum = tl.sum(vals, axis=0)
        vals = vals / (col_sum[None, :] + 1e-6)

    # 3. Store Result
    out_base_ptr = output_ptr + pid * stride_b
    out_ptrs = out_base_ptr + (offs_n[:, None] * stride_n1) + (offs_n[None, :] * stride_n2)
    tl.store(out_ptrs, vals)

# ==========================================
# 2. AUTOGRAD FUNCTION (RECOMPUTE STRATEGY)
# ==========================================
# To save VRAM, we do not store the intermediate Sinkhorn matrices.
# We recompute the manifold projection during the backward pass.

class SinkhornFunction(torch.autograd.Function):
    @staticmethod
    def forward(ctx, logits, n_iters=20):
        # Ensure contiguous memory layout for Triton
        original_shape = logits.shape
        logits_flat = logits.contiguous().view(-1, original_shape[-2], original_shape[-1])
        B_total, N, _ = logits_flat.shape

        output = torch.empty_like(logits_flat)

        # Launch Kernel
        grid = (B_total, )
        sinkhorn_fwd_kernel[grid](
            logits_flat, output,
            logits_flat.stride(0), logits_flat.stride(1), logits_flat.stride(2),
            N=N, ITER=n_iters, BLOCK_SIZE=32
        )

        ctx.save_for_backward(logits)
        ctx.n_iters = n_iters
        return output.view(original_shape)

    @staticmethod
    def backward(ctx, grad_output):
        logits, = ctx.saved_tensors
        n_iters = ctx.n_iters

        with torch.enable_grad():
            x = logits.detach().requires_grad_(True)
            # Re-implement Sinkhorn logic in PyTorch for gradient tracking
            x_stable = x - x.max(dim=-1, keepdim=True)[0].max(dim=-2, keepdim=True)[0].detach()
            M = torch.exp(x_stable)

            for _ in range(n_iters):
                M = M / (M.sum(dim=-1, keepdim=True) + 1e-6)
                M = M / (M.sum(dim=-2, keepdim=True) + 1e-6)

            torch.autograd.backward(M, grad_output)

        return x.grad, None

def sinkhorn_knopp(logits, n_iters=20):
    return SinkhornFunction.apply(logits, n_iters)

# ==========================================
# 3. mHC MODULE DEFINITION
# ==========================================

class ManifoldConstrainedHyperConnection(nn.Module):
    def __init__(self, d_model, expansion_rate=4):
        super().__init__()
        self.d = d_model
        self.n = expansion_rate
        self.input_dim = self.n * self.d

        # Predictors for H_pre (n), H_post (n), H_res (n*n)
        total_pred_params = self.n + self.n + (self.n * self.n)

        # Paper Eq 7: Norm before prediction
        self.stream_norm = nn.RMSNorm(self.input_dim)
        self.predictor = nn.Linear(self.input_dim, total_pred_params)

        # Gating factors (small initial values)
        self.alpha_pre = nn.Parameter(torch.tensor(0.01))
        self.alpha_post = nn.Parameter(torch.tensor(0.01))
        self.alpha_res = nn.Parameter(torch.tensor(0.01))

        # Initialization
        nn.init.normal_(self.predictor.weight, std=0.001) # Small weights
        nn.init.zeros_(self.predictor.bias)

        # BIAS HACK: Force H_res to be Identity Matrix at initialization
        # This prevents the "Loss Explosion" by preserving the residual path.
        res_start = 2 * self.n
        res_end = res_start + (self.n * self.n)
        with torch.no_grad():
            identity_bias = torch.eye(self.n).flatten() * 20.0 # Large value pushes Softmax/Sinkhorn to 1.0
            self.predictor.bias[res_start:res_end] = identity_bias

    def forward(self, X_stream, layer_fn):
        B, S, n, d = X_stream.shape
        X_flat = X_stream.view(B, S, -1)

        # 1. Predict mixing matrices
        X_norm = self.stream_norm(X_flat)
        logits = self.predictor(X_norm)

        h_pre_raw = logits[..., :n]
        h_post_raw = logits[..., n:2*n]
        h_res_raw = logits[..., 2*n:].view(B, S, n, n)

        # 2. Apply Manifold Constraints
        H_pre = torch.sigmoid(h_pre_raw) * self.alpha_pre
        H_post = 2.0 * torch.sigmoid(h_post_raw) * self.alpha_post

        # The Core Innovation: Sinkhorn Projection
        H_res = sinkhorn_knopp(h_res_raw * self.alpha_res)

        # 3. Collapse Stream -> Layer Input
        X_layer_in = torch.einsum('bsn, bsnd -> bsd', H_pre, X_stream)

        # 4. Apply Layer (Attention/FFN)
        X_layer_out = layer_fn(X_layer_in)

        # 5. Expand & Update Stream
        X_mixed = torch.einsum('bsnm, bsmd -> bsnd', H_res, X_stream)
        X_update = torch.einsum('bsn, bsd -> bsnd', H_post, X_layer_out)

        return X_mixed + X_update

# ==========================================
# 4. LLAMA-3 COMPONENTS & ARCHITECTURE
# ==========================================

class LlamaRMSNorm(nn.Module):
    def __init__(self, hidden_size, eps=1e-6):
        super().__init__()
        self.weight = nn.Parameter(torch.ones(hidden_size))
        self.eps = eps
    def forward(self, x):
        var = x.pow(2).mean(-1, keepdim=True)
        return x * torch.rsqrt(var + self.eps) * self.weight

class RotaryEmbedding(nn.Module):
    def __init__(self, dim, max_seq_len=2048):
        super().__init__()
        inv_freq = 1.0 / (10000 ** (torch.arange(0, dim, 2).float() / dim))
        t = torch.arange(max_seq_len).float()
        freqs = torch.outer(t, inv_freq)
        emb = torch.cat((freqs, freqs), dim=-1)
        self.register_buffer("cos", emb.cos()[None, None, :, :], persistent=False)
        self.register_buffer("sin", emb.sin()[None, None, :, :], persistent=False)
    def forward(self, x, seq_len):
        return self.cos[:, :, :seq_len, ...], self.sin[:, :, :seq_len, ...]

def rotate_half(x):
    x1, x2 = x.chunk(2, dim=-1)
    return torch.cat((-x2, x1), dim=-1)

class LlamaAttention(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.n_heads = config['n_heads']
        self.head_dim = config['d_model'] // self.n_heads
        self.wq = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.wk = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.wv = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.wo = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.rope = RotaryEmbedding(self.head_dim, config['context_len'])
    def forward(self, x):
        B, S, D = x.shape
        xq = self.wq(x).view(B, S, self.n_heads, self.head_dim).transpose(1, 2)
        xk = self.wk(x).view(B, S, self.n_heads, self.head_dim).transpose(1, 2)
        xv = self.wv(x).view(B, S, self.n_heads, self.head_dim).transpose(1, 2)
        cos, sin = self.rope(xv, S)
        xq, xk = (xq * cos) + (rotate_half(xq) * sin), (xk * cos) + (rotate_half(xk) * sin)
        out = F.scaled_dot_product_attention(xq, xk, xv, is_causal=True)
        return self.wo(out.transpose(1, 2).contiguous().view(B, S, D))

class LlamaSwiGLUMLP(nn.Module):
    def __init__(self, config):
        super().__init__()
        hid = int(2 * (4 * config['d_model']) / 3)
        self.w1 = nn.Linear(config['d_model'], hid, bias=False)
        self.w2 = nn.Linear(config['d_model'], hid, bias=False)
        self.w3 = nn.Linear(hid, config['d_model'], bias=False)
    def forward(self, x):
        return self.w3(F.silu(self.w1(x)) * self.w2(x))

class MHCLlamaBlock(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.norm1, self.norm2 = LlamaRMSNorm(config['d_model']), LlamaRMSNorm(config['d_model'])
        self.attn, self.mlp = LlamaAttention(config), LlamaSwiGLUMLP(config)
        self.mhc_attn = ManifoldConstrainedHyperConnection(config['d_model'], config['expansion_rate'])
        self.mhc_mlp = ManifoldConstrainedHyperConnection(config['d_model'], config['expansion_rate'])
    def forward(self, stream):
        stream = self.mhc_attn(stream, lambda x: self.attn(self.norm1(x)))
        stream = self.mhc_mlp(stream, lambda x: self.mlp(self.norm2(x)))
        return stream

class NanoLlamaMHC(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.n = config['expansion_rate']
        self.token_emb = nn.Embedding(config['vocab_size'], config['d_model'])
        self.layers = nn.ModuleList([MHCLlamaBlock(config) for _ in range(config['n_layers'])])
        self.norm_final = LlamaRMSNorm(config['d_model'])
        self.lm_head = nn.Linear(config['d_model'], config['vocab_size'], bias=False)
        self.lm_head.weight = self.token_emb.weight
    def forward(self, idx, targets=None):
        x = self.token_emb(idx)
        # Expand stream [B, S, D] -> [B, S, n, D]
        stream = x.unsqueeze(2).expand(-1, -1, self.n, -1).clone()
        for layer in self.layers: stream = layer(stream)
        # Collapse and Output
        logits = self.lm_head(self.norm_final(stream.mean(dim=2)))
        loss = None
        if targets is not None:
            loss = F.cross_entropy(logits.reshape(-1, logits.size(-1)), targets.reshape(-1))
        return logits, loss

# ==========================================
# 5. TRAINING LOOP
# ==========================================

def train_experiment():
    print("Preparing Environment...")
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # 1. Config
    config = {
        'vocab_size': 50257, 'd_model': 256, 'n_layers': 4, 'n_heads': 4,
        'expansion_rate': 4, 'context_len': 128, 'batch_size': 32,
        'lr': 6e-4, 'max_steps': 500
    }

    # 2. Data
    print("Loading Wikitext-2...")
    ds = load_dataset("wikitext", "wikitext-2-raw-v1")
    tok = GPT2Tokenizer.from_pretrained("gpt2"); tok.pad_token = tok.eos_token
    def tokenize(x):
        return tok(x["text"], truncation=True, max_length=config['context_len'], return_overflowing_tokens=True)

    tokenized = ds["train"].map(tokenize, batched=True, remove_columns=ds["train"].column_names)
    tokenized.set_format("pt")

    # Filter short sequences
    valid_indices = [i for i, ids in enumerate(tokenized['input_ids']) if len(ids) == config['context_len']]
    train_set = torch.utils.data.Subset(tokenized, valid_indices)
    loader = DataLoader(train_set, batch_size=config['batch_size'], shuffle=True)

    # 3. Model & Opt
    print(f"Initializing mHC Model (n={config['expansion_rate']})...")
    model = NanoLlamaMHC(config).to(device)
    opt = torch.optim.AdamW(model.parameters(), lr=config['lr'], weight_decay=0.01)

    # Warmup Scheduler
    scheduler = LambdaLR(opt, lambda s: min(s / 50.0, 1.0)) # 50 step warmup

    # 4. Loop
    print("Starting Training...")
    losses, steps = [], []
    step = 0
    model.train()

    while step < config['max_steps']:
        for batch in loader:
            inp = batch['input_ids'].to(device)
            if inp.size(1) != config['context_len']: continue

            x, y = inp[:, :-1], inp[:, 1:]
            opt.zero_grad()
            _, loss = model(x, y)
            loss.backward()

            # Clip gradients to prevent explosion during early Sinkhorn stabilization
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

            opt.step(); scheduler.step()
            step += 1

            if step % 25 == 0:
                print(f"Step {step:03d} | Loss: {loss.item():.4f} | PPL: {math.exp(loss.item()):.2f} | LR: {scheduler.get_last_lr()[0]:.2e}")
                losses.append(loss.item()); steps.append(step)

            if step >= config['max_steps']: break

    # 5. Results
    plt.plot(steps, losses, label='mHC Loss'); plt.grid(True)
    plt.axhline(y=10.8, color='r', linestyle='--', label='Random Baseline')
    plt.title('Rigorous mHC Convergence Test'); plt.legend(); plt.show()
    print("Experiment Complete.")

if __name__ == "__main__":
    train_experiment()
```

```python
# ==============================================================================
# RIGOROUS IMPLEMENTATION: MANIFOLD-CONSTRAINED HYPER-CONNECTIONS (mHC)
#
# Paper Reference: "mHC: Manifold-Constrained Hyper-Connections" (DeepSeek-AI)
# ==============================================================================

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torch.optim.lr_scheduler import LambdaLR
import triton
import triton.language as tl
import math
import time
import matplotlib.pyplot as plt
from transformers import GPT2Tokenizer
from datasets import load_dataset
from itertools import chain

# ==========================================
# 1. TRITON KERNEL: SINKHORN-KNOPP
# ==========================================
@triton.jit
def sinkhorn_fwd_kernel(
    logits_ptr,      # [Total_Tokens, N, N]
    output_ptr,      # [Total_Tokens, N, N]
    stride_b,        # Stride for batch dimension
    stride_n1,       # Stride for first N
    stride_n2,       # Stride for second N
    N: tl.constexpr, # Expansion rate (e.g., 4)
    ITER: tl.constexpr,
    BLOCK_SIZE: tl.constexpr
):
    pid = tl.program_id(0)
    offs_n = tl.arange(0, N)

    # Calculate pointer offsets
    base_ptr = logits_ptr + pid * stride_b
    ptrs = base_ptr + (offs_n[:, None] * stride_n1) + (offs_n[None, :] * stride_n2)

    # Load Logits
    vals = tl.load(ptrs)

    # 1. Exponentiate (Stability: subtract max)
    max_val = tl.max(vals)
    vals = tl.exp(vals - max_val)

    # 2. Sinkhorn Iterations (In-Register)
    for i in range(ITER):
        # Row Norm
        row_sum = tl.sum(vals, axis=1)
        vals = vals / (row_sum[:, None] + 1e-6)
        # Column Norm
        col_sum = tl.sum(vals, axis=0)
        vals = vals / (col_sum[None, :] + 1e-6)

    # 3. Store Result
    out_base_ptr = output_ptr + pid * stride_b
    out_ptrs = out_base_ptr + (offs_n[:, None] * stride_n1) + (offs_n[None, :] * stride_n2)
    tl.store(out_ptrs, vals)

# ==========================================
# 2. AUTOGRAD FUNCTION (RECOMPUTE STRATEGY)
# ==========================================
class SinkhornFunction(torch.autograd.Function):
    @staticmethod
    def forward(ctx, logits, n_iters=20):
        # Ensure contiguous memory layout for Triton
        original_shape = logits.shape
        logits_flat = logits.contiguous().view(-1, original_shape[-2], original_shape[-1])
        B_total, N, _ = logits_flat.shape

        output = torch.empty_like(logits_flat)

        # Launch Kernel
        grid = (B_total, )
        sinkhorn_fwd_kernel[grid](
            logits_flat, output,
            logits_flat.stride(0), logits_flat.stride(1), logits_flat.stride(2),
            N=N, ITER=n_iters, BLOCK_SIZE=32
        )

        ctx.save_for_backward(logits)
        ctx.n_iters = n_iters
        return output.view(original_shape)

    @staticmethod
    def backward(ctx, grad_output):
        logits, = ctx.saved_tensors
        n_iters = ctx.n_iters

        with torch.enable_grad():
            x = logits.detach().requires_grad_(True)
            # Re-implement Sinkhorn logic in PyTorch for gradient tracking
            x_stable = x - x.max(dim=-1, keepdim=True)[0].max(dim=-2, keepdim=True)[0].detach()
            M = torch.exp(x_stable)

            for _ in range(n_iters):
                M = M / (M.sum(dim=-1, keepdim=True) + 1e-6)
                M = M / (M.sum(dim=-2, keepdim=True) + 1e-6)

            torch.autograd.backward(M, grad_output)

        return x.grad, None

def sinkhorn_knopp(logits, n_iters=20):
    return SinkhornFunction.apply(logits, n_iters)

# ==========================================
# 3. mHC MODULE DEFINITION
# ==========================================
class ManifoldConstrainedHyperConnection(nn.Module):
    def __init__(self, d_model, expansion_rate=4):
        super().__init__()
        self.d = d_model
        self.n = expansion_rate
        self.input_dim = self.n * self.d

        # Predictors for H_pre (n), H_post (n), H_res (n*n)
        total_pred_params = self.n + self.n + (self.n * self.n)

        # Paper Eq 7: Norm before prediction
        self.stream_norm = nn.RMSNorm(self.input_dim)
        self.predictor = nn.Linear(self.input_dim, total_pred_params)

        # Gating factors (small initial values)
        self.alpha_pre = nn.Parameter(torch.tensor(0.01))
        self.alpha_post = nn.Parameter(torch.tensor(0.01))
        self.alpha_res = nn.Parameter(torch.tensor(0.01))

        # Initialization
        nn.init.normal_(self.predictor.weight, std=0.001)
        nn.init.zeros_(self.predictor.bias)

        # BIAS HACK: Force H_res to be Identity Matrix at initialization
        # This prevents the "Loss Explosion"
        res_start = 2 * self.n
        res_end = res_start + (self.n * self.n)
        with torch.no_grad():
            identity_bias = torch.eye(self.n).flatten() * 20.0
            self.predictor.bias[res_start:res_end] = identity_bias

    def forward(self, X_stream, layer_fn):
        B, S, n, d = X_stream.shape
        X_flat = X_stream.view(B, S, -1)

        # 1. Predict mixing matrices
        X_norm = self.stream_norm(X_flat)
        logits = self.predictor(X_norm)

        h_pre_raw = logits[..., :n]
        h_post_raw = logits[..., n:2*n]
        h_res_raw = logits[..., 2*n:].view(B, S, n, n)

        # 2. Apply Manifold Constraints
        H_pre = torch.sigmoid(h_pre_raw) * self.alpha_pre
        H_post = 2.0 * torch.sigmoid(h_post_raw) * self.alpha_post
        H_res = sinkhorn_knopp(h_res_raw * self.alpha_res)

        # 3. Collapse Stream -> Layer Input
        X_layer_in = torch.einsum('bsn, bsnd -> bsd', H_pre, X_stream)

        # 4. Apply Layer (Attention/FFN)
        X_layer_out = layer_fn(X_layer_in)

        # 5. Expand & Update Stream
        X_mixed = torch.einsum('bsnm, bsmd -> bsnd', H_res, X_stream)
        X_update = torch.einsum('bsn, bsd -> bsnd', H_post, X_layer_out)

        return X_mixed + X_update

# ==========================================
# 4. LLAMA-3 ARCHITECTURE
# ==========================================
class LlamaRMSNorm(nn.Module):
    def __init__(self, hidden_size, eps=1e-6):
        super().__init__()
        self.weight = nn.Parameter(torch.ones(hidden_size))
        self.eps = eps
    def forward(self, x):
        var = x.pow(2).mean(-1, keepdim=True)
        return x * torch.rsqrt(var + self.eps) * self.weight

class RotaryEmbedding(nn.Module):
    def __init__(self, dim, max_seq_len=2048):
        super().__init__()
        inv_freq = 1.0 / (10000 ** (torch.arange(0, dim, 2).float() / dim))
        t = torch.arange(max_seq_len).float()
        freqs = torch.outer(t, inv_freq)
        emb = torch.cat((freqs, freqs), dim=-1)
        self.register_buffer("cos", emb.cos()[None, None, :, :], persistent=False)
        self.register_buffer("sin", emb.sin()[None, None, :, :], persistent=False)
    def forward(self, x, seq_len):
        return self.cos[:, :, :seq_len, ...], self.sin[:, :, :seq_len, ...]

def rotate_half(x):
    x1, x2 = x.chunk(2, dim=-1)
    return torch.cat((-x2, x1), dim=-1)

class LlamaAttention(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.n_heads = config['n_heads']
        self.head_dim = config['d_model'] // self.n_heads
        self.wq = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.wk = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.wv = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.wo = nn.Linear(config['d_model'], config['d_model'], bias=False)
        self.rope = RotaryEmbedding(self.head_dim, config['context_len'])
    def forward(self, x):
        B, S, D = x.shape
        xq = self.wq(x).view(B, S, self.n_heads, self.head_dim).transpose(1, 2)
        xk = self.wk(x).view(B, S, self.n_heads, self.head_dim).transpose(1, 2)
        xv = self.wv(x).view(B, S, self.n_heads, self.head_dim).transpose(1, 2)
        cos, sin = self.rope(xv, S)
        xq, xk = (xq * cos) + (rotate_half(xq) * sin), (xk * cos) + (rotate_half(xk) * sin)
        out = F.scaled_dot_product_attention(xq, xk, xv, is_causal=True)
        return self.wo(out.transpose(1, 2).contiguous().view(B, S, D))

class LlamaSwiGLUMLP(nn.Module):
    def __init__(self, config):
        super().__init__()
        hid = int(2 * (4 * config['d_model']) / 3)
        self.w1 = nn.Linear(config['d_model'], hid, bias=False)
        self.w2 = nn.Linear(config['d_model'], hid, bias=False)
        self.w3 = nn.Linear(hid, config['d_model'], bias=False)
    def forward(self, x):
        return self.w3(F.silu(self.w1(x)) * self.w2(x))

class MHCLlamaBlock(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.norm1, self.norm2 = LlamaRMSNorm(config['d_model']), LlamaRMSNorm(config['d_model'])
        self.attn, self.mlp = LlamaAttention(config), LlamaSwiGLUMLP(config)
        self.mhc_attn = ManifoldConstrainedHyperConnection(config['d_model'], config['expansion_rate'])
        self.mhc_mlp = ManifoldConstrainedHyperConnection(config['d_model'], config['expansion_rate'])
    def forward(self, stream):
        stream = self.mhc_attn(stream, lambda x: self.attn(self.norm1(x)))
        stream = self.mhc_mlp(stream, lambda x: self.mlp(self.norm2(x)))
        return stream

class NanoLlamaMHC(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.n = config['expansion_rate']
        self.token_emb = nn.Embedding(config['vocab_size'], config['d_model'])
        self.layers = nn.ModuleList([MHCLlamaBlock(config) for _ in range(config['n_layers'])])
        self.norm_final = LlamaRMSNorm(config['d_model'])
        self.lm_head = nn.Linear(config['d_model'], config['vocab_size'], bias=False)
        self.lm_head.weight = self.token_emb.weight
    def forward(self, idx, targets=None):
        x = self.token_emb(idx)
        # Expand stream [B, S, D] -> [B, S, n, D]
        stream = x.unsqueeze(2).expand(-1, -1, self.n, -1).clone()
        for layer in self.layers: stream = layer(stream)
        # Collapse and Output
        logits = self.lm_head(self.norm_final(stream.mean(dim=2)))
        loss = None
        if targets is not None:
            # FIX: reshape for non-contiguous views
            loss = F.cross_entropy(logits.reshape(-1, logits.size(-1)), targets.reshape(-1))
        return logits, loss

# ==========================================
# 5. ROBUST TRAINING LOOP
# ==========================================

def train_experiment():
    print("Preparing Environment...")
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Config
    config = {
        'vocab_size': 50257, 'd_model': 256, 'n_layers': 4, 'n_heads': 4,
        'expansion_rate': 4, 'context_len': 128, 'batch_size': 32,
        'lr': 6e-4, 'max_steps': 600
    }

    # DATA LOADING: Concatenate & Chunk Method (Fixes Jagged Tensor Errors)
    print("Loading Wikitext-2...")
    ds = load_dataset("wikitext", "wikitext-2-raw-v1")
    tok = GPT2Tokenizer.from_pretrained("gpt2"); tok.pad_token = tok.eos_token

    def tokenize_function(examples):
        return tok(examples["text"])

    print("Tokenizing...")
    tokenized_ds = ds["train"].map(tokenize_function, batched=True, remove_columns=["text"])

    block_size = config['context_len']
    def group_texts(examples):
        # Concatenate all texts
        concatenated = {k: list(chain(*examples[k])) for k in examples.keys()}
        total_length = len(concatenated[list(examples.keys())[0]])
        # Drop remainder
        if total_length >= block_size:
            total_length = (total_length // block_size) * block_size
        # Split into blocks
        result = {
            k: [t[i : i + block_size] for i in range(0, total_length, block_size)]
            for k, t in concatenated.items()
        }
        return result

    print("Chunking...")
    lm_ds = tokenized_ds.map(group_texts, batched=True)
    lm_ds.set_format("pt")

    loader = DataLoader(lm_ds, batch_size=config['batch_size'], shuffle=True)

    # Model Setup
    print(f"Initializing mHC Model (n={config['expansion_rate']})...")
    model = NanoLlamaMHC(config).to(device)
    opt = torch.optim.AdamW(model.parameters(), lr=config['lr'], weight_decay=0.01)

    # Warmup
    scheduler = LambdaLR(opt, lambda s: min(s / 50.0, 1.0))

    # Train
    print("Starting Training...")
    losses, steps = [], []
    step = 0
    model.train()

    # Loop over epochs as needed to hit max_steps
    while step < config['max_steps']:
        for batch in loader:
            inp = batch['input_ids'].to(device)
            if inp.size(1) != config['context_len']: continue

            x, y = inp[:, :-1], inp[:, 1:]

            opt.zero_grad()
            _, loss = model(x, y)
            loss.backward()

            # Clip Gradients
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

            opt.step(); scheduler.step()
            step += 1

            if step % 25 == 0:
                print(f"Step {step:03d} | Loss: {loss.item():.4f} | PPL: {math.exp(loss.item()):.2f} | LR: {scheduler.get_last_lr()[0]:.2e}")
                losses.append(loss.item()); steps.append(step)

            if step >= config['max_steps']: break

    # Plot
    plt.plot(steps, losses, label='mHC Loss'); plt.grid(True)
    plt.axhline(y=10.82, color='r', linestyle='--', label='Random Chance')
    plt.title('Rigorous mHC Convergence Test'); plt.legend(); plt.show()
    print("Experiment Complete.")

if __name__ == "__main__":
    train_experiment()
```
