```python
import torch
import math
from dataclasses import dataclass, field
from typing import Optional, Tuple, List, Iterator
from transformers import AutoTokenizer
from datasets import load_dataset
from torch.utils.data import IterableDataset, DataLoader

# ==========================================
# Component 1: The Master Configuration
# ==========================================

@dataclass
class SFConfig:
    """
    Configuration for Skeleton & Flesh (S&F) Training.
    Defaults set for a ~350M parameter model on A100.
    """
    # --- Model Architecture (Llama-style 350M) ---
    vocab_size: int = 50304  # Standard GPT-2 size (padded for multiples of 8)
    dim: int = 1024
    n_layers: int = 24
    n_heads: int = 16
    n_kv_heads: int = 16  # Multi-head attention (can be < n_heads for GQA)
    head_dim: int = 64
    max_seq_len: int = 1024
    dropout: float = 0.0

    # --- Training Hyperparameters ---
    batch_size: int = 32      # Global batch size
    grad_accumulation_steps: int = 4
    learning_rate: float = 3e-4
    weight_decay: float = 0.1
    beta1: float = 0.9
    beta2: float = 0.95
    max_grad_norm: float = 1.0

    # --- WSD Scheduler & S&F Dynamics ---
    total_steps: int = 10000  # Total training steps (scaled down for demo, normally 100k+)
    warmup_ratio: float = 0.1 # 10% Warmup
    stable_ratio: float = 0.8 # 80% Stable Phase (Where S&F happens)
    decay_ratio: float = 0.1  # 10% Decay (Annealing)

    # --- Sparsity & Regrowth ---
    target_density: float = 0.2        # Final model should be 20% dense
    update_interval: int = 1000        # How often to perform Surgery (steps)
    pma_buffer_size: int = 10          # Number of checkpoints for SNR calculation
    snr_prune_rate: float = 0.3        # Prune bottom 30% of weights every cycle
    flesh_boost_factor: float = 10.0   # 10x LR for new weights
    flesh_boost_steps: int = 200       # Duration of boost

    @property
    def device(self):
        return "cuda" if torch.cuda.is_available() else "cpu"

    @property
    def dtype(self):
        return torch.bfloat16 if torch.cuda.is_bf16_supported() else torch.float16

# ==========================================
# Component 2: The Production Data Pipeline
# ==========================================

class StreamingPackedDataset(IterableDataset):
    """
    A robust streaming dataset that:
    1. Streams text from C4 (infinite).
    2. Tokenizes on the fly.
    3. 'Packs' tokens into fixed chunks of max_seq_len.
    4. Handles buffer shuffling for randomness.
    """
    def __init__(self, tokenizer, seq_len: int, buffer_size: int = 10000):
        self.tokenizer = tokenizer
        self.seq_len = seq_len
        self.buffer_size = buffer_size
        # Load C4 (en) in streaming mode.
        # We use 'allenai/c4' -> 'en'.
        self.dataset = load_dataset("allenai/c4", "en", split="train", streaming=True)

    def __iter__(self):
        iterator = iter(self.dataset)
        more_examples = True
        token_buffer = []

        while more_examples:
            # 1. Fill the buffer with raw text tokens
            try:
                # Fetch a chunk of examples to reduce overhead
                for _ in range(100):
                    data = next(iterator)
                    text = data['text']
                    # Tokenize and append EOS token
                    tokens = self.tokenizer(text, truncation=False, add_special_tokens=False)['input_ids']
                    tokens.append(self.tokenizer.eos_token_id)
                    token_buffer.extend(tokens)
            except StopIteration:
                more_examples = False

            # 2. Yield fixed-size chunks
            while len(token_buffer) >= self.seq_len:
                chunk = token_buffer[:self.seq_len]
                token_buffer = token_buffer[self.seq_len:]

                # Input is chunk, Target is shifted chunk (handled in loss, but we yield x, y here)
                # Ideally, for pretraining, we return x=chunk. The shifting happens inside the model forward.
                yield torch.tensor(chunk, dtype=torch.long)

# Helper to verify the pipeline
def create_dataloader(config: SFConfig):
    print("Loading Tokenizer (GPT-2 standard)...")
    tokenizer = AutoTokenizer.from_pretrained("gpt2")
    tokenizer.pad_token = tokenizer.eos_token # Fix for GPT-2

    print("Initializing Streaming Pipeline...")
    dataset = StreamingPackedDataset(tokenizer, config.max_seq_len)

    dataloader = DataLoader(
        dataset,
        batch_size=config.batch_size,
        pin_memory=True,
        num_workers=2 # Adjust based on CPU cores
    )
    return dataloader, tokenizer

# ==========================================
# VERIFICATION BLOCK
# ==========================================
if __name__ == "__main__":
    # Initialize Config
    config = SFConfig()
    print(f"Configuration Loaded. Device: {config.device}, Dtype: {config.dtype}")

    # Initialize Data
    train_loader, tokenizer = create_dataloader(config)

    # Sanity Check: Fetch one batch
    print("Fetching first batch to verify shape and integrity...")
    try:
        batch = next(iter(train_loader))
        print(f"Batch Shape: {batch.shape}")
        print(f"Data Type: {batch.dtype}")
        print(f"Sample Token IDs: {batch[0, :10]}")
        print("Data Pipeline: READY")
    except Exception as e:
        print(f"Data Pipeline Failed: {e}")
```

```python
import torch.nn as nn
import torch.nn.functional as F
from typing import Dict, List, Optional
import copy

# ==========================================
# Component 3: The SparseLlama Architecture
# ==========================================

class RMSNorm(nn.Module):
    def __init__(self, dim: int, eps: float = 1e-6):
        super().__init__()
        self.eps = eps
        self.weight = nn.Parameter(torch.ones(dim))

    def forward(self, x):
        var = torch.mean(x ** 2, dim=-1, keepdim=True)
        x_normed = x * torch.rsqrt(var + self.eps)
        return self.weight * x_normed

def apply_rotary_emb(x, dim):
    # Simplified RoPE implementation for brevity/speed
    # x: [Batch, Seq, Head, Dim]
    half_dim = dim // 2
    x1, x2 = x[..., :half_dim], x[..., half_dim:]
    # Ideally we precompute cos/sin frequencies, but calculating on fly
    # for this scale is acceptable overhead for readability.
    seq_len = x.shape[1]
    device = x.device
    inv_freq = 1.0 / (10000 ** (torch.arange(0, dim, 2, device=device).float() / dim))
    t = torch.arange(seq_len, device=device).type_as(inv_freq)
    freqs = torch.einsum('i,j->ij', t, inv_freq)
    emb = torch.cat((freqs, freqs), dim=-1).to(x.dtype)

    # Rotation
    cos = emb.cos()[None, :, None, :]
    sin = emb.sin()[None, :, None, :]
    return (x * cos) + (torch.cat((-x2, x1), dim=-1) * sin)

class SFLayer(nn.Linear):
    """
    Skeleton & Flesh Layer.
    Standard Linear layer, but holds a persistent binary mask.
    Includes hooks to compute dense gradients for 'Flesh' scouting.
    """
    def __init__(self, in_features, out_features, bias=False):
        super().__init__(in_features, out_features, bias=bias)
        # Register mask as a buffer (persistent state, not a parameter)
        self.register_buffer("mask", torch.ones_like(self.weight))

        # Buffers for Gradient Scouting (Regrowth)
        self._scout_input = None
        self._scout_grad_output = None
        self._scouting_active = False
        self._fhooks = []
        self._bhooks = []

    def forward(self, x):
        # Apply mask during forward pass
        # Note: We use weight * mask. Autograd will naturally give 0 grad
        # for masked weights, which is correct for standard training.
        w_masked = self.weight * self.mask
        return F.linear(x, w_masked, self.bias)

    def enable_gradient_scouting(self):
        """Attaches hooks to capture inputs and grad_outputs for dense gradient calculation."""
        self._scouting_active = True

        def save_input_hook(module, input, output):
            # Input is a tuple, take [0]. Detach to save graph memory, we only need values.
            self._scout_input = input[0].detach()

        def save_grad_output_hook(module, grad_input, grad_output):
            # grad_output is a tuple.
            self._scout_grad_output = grad_output[0].detach()

        self._fhooks.append(self.register_forward_hook(save_input_hook))
        self._bhooks.append(self.register_full_backward_hook(save_grad_output_hook))

    def disable_gradient_scouting(self):
        """Removes hooks and clears buffers."""
        self._scouting_active = False
        for h in self._fhooks: h.remove()
        for h in self._bhooks: h.remove()
        self._fhooks = []
        self._bhooks = []
        self._scout_input = None
        self._scout_grad_output = None

    def compute_dense_grad(self):
        """
        Manually computes G = (grad_output^T @ input)
        Returns the dense gradient tensor for regrowth analysis.
        """
        if self._scout_input is None or self._scout_grad_output is None:
            return None

        # Inputs: [Batch, Seq, InDim]
        # Grads:  [Batch, Seq, OutDim]
        # We need to flatten batch/seq dimensions

        # x: [N, In]
        x = self._scout_input.reshape(-1, self.in_features)
        # gy: [N, Out]
        gy = self._scout_grad_output.reshape(-1, self.out_features)

        # Dense Grad = gy.T @ x -> [Out, In]
        # We assume BF16, doing math in Float32 for precision if needed,
        # but BF16 should suffice for selection.
        dense_grad = torch.matmul(gy.t(), x)
        return dense_grad

class CausalSelfAttention(nn.Module):
    def __init__(self, config: SFConfig):
        super().__init__()
        assert config.dim % config.n_heads == 0
        self.n_heads = config.n_heads
        self.head_dim = config.dim // config.n_heads

        # S&F Layers for Projections
        self.q_proj = SFLayer(config.dim, config.dim)
        self.k_proj = SFLayer(config.dim, config.dim)
        self.v_proj = SFLayer(config.dim, config.dim)
        self.o_proj = SFLayer(config.dim, config.dim)

        self.dropout = config.dropout

    def forward(self, x):
        B, T, C = x.size()

        q = self.q_proj(x).view(B, T, self.n_heads, self.head_dim)
        k = self.k_proj(x).view(B, T, self.n_heads, self.head_dim)
        v = self.v_proj(x).view(B, T, self.n_heads, self.head_dim)

        # Apply RoPE
        q = apply_rotary_emb(q, self.head_dim)
        k = apply_rotary_emb(k, self.head_dim)

        # Flash Attention (SDPA)
        q = q.transpose(1, 2) # [B, H, T, D]
        k = k.transpose(1, 2)
        v = v.transpose(1, 2)

        out = F.scaled_dot_product_attention(q, k, v, is_causal=True, dropout_p=self.dropout if self.training else 0)

        out = out.transpose(1, 2).contiguous().view(B, T, C)
        return self.o_proj(out)

class MLP(nn.Module):
    def __init__(self, config: SFConfig):
        super().__init__()
        hidden_dim = 4 * config.dim
        hidden_dim = int(2 * hidden_dim / 3) # SwiGLU sizing

        self.gate_proj = SFLayer(config.dim, hidden_dim)
        self.up_proj = SFLayer(config.dim, hidden_dim)
        self.down_proj = SFLayer(hidden_dim, config.dim)

    def forward(self, x):
        return self.down_proj(F.silu(self.gate_proj(x)) * self.up_proj(x))

class SparseLlamaBlock(nn.Module):
    def __init__(self, config: SFConfig):
        super().__init__()
        self.ln1 = RMSNorm(config.dim)
        self.attn = CausalSelfAttention(config)
        self.ln2 = RMSNorm(config.dim)
        self.mlp = MLP(config)

    def forward(self, x):
        x = x + self.attn(self.ln1(x))
        x = x + self.mlp(self.ln2(x))
        return x

class SparseLlama(nn.Module):
    def __init__(self, config: SFConfig):
        super().__init__()
        self.config = config
        self.embed = nn.Embedding(config.vocab_size, config.dim)
        self.layers = nn.ModuleList([SparseLlamaBlock(config) for _ in range(config.n_layers)])
        self.ln_f = RMSNorm(config.dim)
        self.head = SFLayer(config.dim, config.vocab_size, bias=False)

        # Weight tying
        self.embed.weight = self.head.weight

        self.apply(self._init_weights)

    def _init_weights(self, module):
        if isinstance(module, (nn.Linear, SFLayer)):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)
        elif isinstance(module, nn.Embedding):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)

    def forward(self, x, targets=None):
        x = self.embed(x)
        for layer in self.layers:
            x = layer(x)
        x = self.ln_f(x)
        logits = self.head(x)

        loss = None
        if targets is not None:
            # Flatten for CrossEntropy
            loss = F.cross_entropy(logits.view(-1, logits.size(-1)), targets.view(-1))

        return logits, loss
```

```python
import torch
import torch.nn as nn
from typing import Dict, List, Optional

# ==========================================
# Component 4: The Skeleton Engine (Brains)
# [FIXED: Soft Regrowth Logic]
# ==========================================

class WelfordTracker:
    def __init__(self):
        self.stats = {}

    def update(self, model: nn.Module):
        with torch.no_grad():
            for name, param in model.named_parameters(remove_duplicate=False):
                if not param.requires_grad: continue
                if name not in self.stats:
                    self.stats[name] = {
                        'mean': torch.zeros_like(param),
                        'M2': torch.zeros_like(param),
                        'n': 0
                    }
                s = self.stats[name]
                s['n'] += 1
                delta = param - s['mean']
                s['mean'] += delta / s['n']
                delta2 = param - s['mean']
                s['M2'] += delta * delta2

    def get_snr(self, name: str, epsilon: float = 1e-8):
        if name not in self.stats: return None, None
        s = self.stats[name]
        variance = s['M2'] / s['n']
        snr = torch.abs(s['mean']) / torch.sqrt(variance + epsilon)
        return snr, s['mean']

    def reset(self):
        self.stats = {}

class SkeletonEngine:
    def __init__(self, model: nn.Module, config):
        self.model = model
        self.config = config
        self.tracker = WelfordTracker()
        self.boost_mask = {}
        self.last_pruned_count = 0 # Track how many we killed to know how many to birth

        self.sf_layers = {}
        for name, module in model.named_modules():
            if hasattr(module, "mask"):
                param_name = name + ".weight" if name else "weight"
                self.sf_layers[param_name] = module

    def step_welford(self):
        self.tracker.update(self.model)

    def perform_surgery(self):
        print("\n[SkeletonEngine] Starting Surgery...")

        # 1. Collect SNR
        all_snrs = []
        valid_layers = []
        for name in self.sf_layers:
            snr, _ = self.tracker.get_snr(name)
            if snr is None: continue
            valid_layers.append(name)
            mask = self.sf_layers[name].mask
            # Flatten active SNR scores
            all_snrs.append(snr[mask == 1].view(-1))

        if not all_snrs:
            return "ERROR"

        global_scores = torch.cat(all_snrs)
        k_prune = int(len(global_scores) * self.config.snr_prune_rate)
        k_prune = max(1, min(k_prune, len(global_scores) - 1))

        # Determine threshold
        threshold = torch.topk(global_scores, len(global_scores) - k_prune, largest=True).values[-1]
        print(f"[SkeletonEngine] Pruning weights with SNR < {threshold:.4f}")

        self.last_pruned_count = 0

        # 2. Prune
        for name in valid_layers:
            layer = self.sf_layers[name]
            snr, mean_weight = self.tracker.get_snr(name)

            # Reset Skeleton to Basin Center
            layer.weight.data.copy_(mean_weight)

            # Update Mask (Keep High SNR)
            keep_mask = (snr >= threshold).float()
            new_mask = layer.mask * keep_mask

            pruned = (layer.mask - new_mask).sum().item()
            self.last_pruned_count += pruned

            layer.mask.copy_(new_mask)
            layer.weight.data.mul_(layer.mask) # Zero out

        print(f"[SkeletonEngine] Pruned {int(self.last_pruned_count)} connections.")

        # 3. Request Regrowth
        print("[SkeletonEngine] Enabling Hooks for Gradient Scouting...")
        for layer in self.sf_layers.values():
            layer.enable_gradient_scouting()

        return "REQ_SCOUT"

    def finish_surgery(self):
        total_grown = 0
        self.boost_mask = {}

        all_grads = []
        grad_map = {}

        # Collect Gradients
        for name, layer in self.sf_layers.items():
            dense_grad = layer.compute_dense_grad()
            layer.disable_gradient_scouting()

            if dense_grad is None: continue

            # Candidates: Inactive weights (mask == 0)
            inactive_grads = torch.abs(dense_grad) * (1 - layer.mask)
            grad_map[name] = inactive_grads
            all_grads.append(inactive_grads.view(-1))

        if not all_grads:
            print("[SkeletonEngine] No gradients captured.")
            self.tracker.reset()
            return

        # --- LOGIC UPDATE: SOFT REGROWTH ---
        # We calculate needed slots based on Soft Decay.
        # Strategy: Always regrow 50% of what we just pruned to ensure Flesh dynamics
        # UNLESS we are already below target density.

        flat_grads = torch.cat(all_grads)

        total_params = sum(l.weight.numel() for l in self.sf_layers.values())
        min_target = int(total_params * self.config.target_density)
        current_active = sum(l.mask.sum().item() for l in self.sf_layers.values())

        # Base need to reach minimum floor
        floor_need = min_target - current_active

        # Soft need: Regrow portion of pruned to enable exploration
        # Let's say we regrow 50% of the pruned weights to cycle them
        cycle_need = int(self.last_pruned_count * 0.5)

        # We take the MAX of floor_need and cycle_need
        # This ensures we don't shrink too fast (via cycle_need)
        # And we don't stay too small (via floor_need)
        needed = max(floor_need, cycle_need)

        # Cap needed at available slots
        needed = min(needed, len(flat_grads))

        if needed > 0:
            print(f"[SkeletonEngine] Attempting to regrow {int(needed)} connections (Soft Regrowth)...")

            # Top-K Gradients
            grad_threshold = torch.topk(flat_grads, int(needed)).values[-1]

            for name, layer in self.sf_layers.items():
                if name not in grad_map: continue
                g = grad_map[name]

                flesh_mask = (g >= grad_threshold).float() * (1 - layer.mask)

                # Update Mask
                layer.mask.add_(flesh_mask).clamp_(max=1.0)

                # Zero-Init Flesh
                with torch.no_grad():
                    layer.weight.data.mul_(1.0 - flesh_mask)

                # Register for Boost
                self.boost_mask[layer.weight] = flesh_mask.bool()
                total_grown += flesh_mask.sum().item()

        print(f"[SkeletonEngine] Regrew {int(total_grown)} connections.")
        self.tracker.reset()
```

```python
import torch.optim as optim
import time

# ==========================================
# Component 5: The S&F Training Loop
# [FIXED: Infinite Loop Bug]
# ==========================================

def get_lr(step, config):
    # (Same as before)
    warmup_steps = int(config.total_steps * config.warmup_ratio)
    if step < warmup_steps:
        return config.learning_rate * (step / max(1, warmup_steps))

    decay_start = config.total_steps - int(config.total_steps * config.decay_ratio)
    if step < decay_start:
        return config.learning_rate

    remaining = config.total_steps - step
    total_decay = config.total_steps - decay_start
    return config.learning_rate * (remaining / max(1, total_decay))

def train(model, dataloader, config, engine):
    print("\n[Training] Initializing Optimizer...")
    optimizer = optim.AdamW(
        model.parameters(),
        lr=config.learning_rate,
        betas=(config.beta1, config.beta2),
        weight_decay=config.weight_decay
    )

    model.train()
    global_step = 0
    t0 = time.time()

    boost_remaining = 0

    data_iter = iter(dataloader)

    print(f"[Training] Starting Loop for {config.total_steps} steps...")

    while global_step < config.total_steps:
        # 1. Fetch Data
        try:
            batch = next(data_iter)
        except StopIteration:
            data_iter = iter(dataloader)
            batch = next(data_iter)

        inputs = batch[:, :-1].to(model.config.device)
        targets = batch[:, 1:].to(model.config.device)

        # 2. WSD Schedule
        current_lr = get_lr(global_step, config)
        for param_group in optimizer.param_groups:
            param_group['lr'] = current_lr

        # 3. Determine Phase
        warmup_steps = int(config.total_steps * config.warmup_ratio)
        decay_start = config.total_steps - int(config.total_steps * config.decay_ratio)
        is_stable_phase = (global_step >= warmup_steps) and (global_step < decay_start)

        step_performed = False # Flag to track if we did a standard step

        # 4. Surgery Logic (Only in Stable Phase)
        if is_stable_phase:
            engine.step_welford()

            # Trigger Surgery
            if global_step > 0 and global_step % config.update_interval == 0:
                # -----------------------------
                # SURGERY START
                # -----------------------------
                status = engine.perform_surgery()

                if status == "REQ_SCOUT":
                    # Gradient Scouting Step
                    optimizer.zero_grad()
                    _, loss = model(inputs, targets)
                    loss.backward()

                    # Compute gradients and regrow
                    engine.finish_surgery()

                    boost_remaining = config.flesh_boost_steps
                    print(f"[Training] Surgery Done at Step {global_step}. Boosting Flesh LR.")

                    # We count the scouting pass as the step for this iteration
                    # We do NOT step the optimizer, effectively resetting momentum
                    optimizer.zero_grad()

                    # Log surgery step
                    print(f"Step {global_step} | Type: SURGERY | Density: Check next step")

                    global_step += 1 # CRITICAL FIX: Increment step
                    t0 = time.time() # Reset timer
                    continue
                # -----------------------------
                # SURGERY END
                # -----------------------------

        # 5. Standard Forward Pass
        optimizer.zero_grad()
        logits, loss = model(inputs, targets)
        loss.backward()

        # 6. Gradient Post-Processing
        # A. Mask Gradients
        with torch.no_grad():
            for name, layer in engine.sf_layers.items():
                if layer.weight.grad is not None:
                    layer.weight.grad.mul_(layer.mask)

        # B. Apply Flesh Boost
        if boost_remaining > 0:
            with torch.no_grad():
                for param, mask in engine.boost_mask.items():
                    if param.grad is not None:
                        param.grad.add_(param.grad * mask * (config.flesh_boost_factor - 1))
            boost_remaining -= 1

        # C. Clip Grad Norm
        torch.nn.utils.clip_grad_norm_(model.parameters(), config.max_grad_norm)

        # 7. Step
        optimizer.step()

        # 8. Logging
        if global_step % 10 == 0:
            dt = time.time() - t0
            # Density Check
            active_sum = sum(l.mask.sum().item() for l in engine.sf_layers.values())
            total_sum = sum(l.weight.numel() for l in engine.sf_layers.values())
            density = active_sum / total_sum

            print(f"Step {global_step} | Loss: {loss.item():.4f} | LR: {current_lr:.2e} | Density: {density:.2%} | Boost: {boost_remaining > 0}")
            t0 = time.time()

        global_step += 1

    print("[Training] Completed.")
    return model
```
